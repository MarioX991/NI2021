### Ova sveska približava jedinstvenost audio spektrograma. 

Možete eksperimentisati sa različitim zvučnim zapisima (`sample1.wav`, `sample2.wav`, `sample3.wav` i `sample4.wav`) i uporediti dobijene spektrograme. Ovoga puta su svi zvučni zapisi jednokanalni.  

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
from scipy import signal
from scipy.io import wavfile
from IPython import display
```

```python
audio_path = 'data/sample3.wav'
```

```python
display.Audio(audio_path)
```

```python
Fs, samples = wavfile.read(audio_path)
```

```python
Fs
```

```python
samples.shape
```

```python
f, t, s = signal.spectrogram(samples, Fs)
```

```python
plt.pcolormesh(t, f, np.log(s),shading='auto')
plt.colorbar()
plt.xlabel('Time [sec]')
plt.ylim(0, 2500)
plt.ylabel('Frequency [Hz]')
plt.show()
```
